````r
#' simple function for reading in a diag distance matrix 
read_mtx = function(F){
    cmd = paste0('gunzip -c ', F)
    X = fread(cmd, sep=';') %>%
        as.data.frame %>%
        rename('X' = V1) %>%
        gather(Y, Z, -X)
    return(X)
}
````

#### Bracken % mapped

````r
# listing bracken kreport files
kreports = list.files(file.path(out_dir, 'kraken'), pattern='*_bracken.kreport', full.names=TRUE)
kreports %>% length %>% print
````

````r
# reading in tables
read_kreport = function(x){
    df = read.delim(x, sep='\t', header=FALSE)
    df$Sample = gsub('.+/|_bracken.kreport', '', x)
    return(df)
}

registerDoParallel(threads)
df_kr = plyr::llply(as.list(kreports), read_kreport, .parallel=TRUE) %>%
    do.call(rbind, .) %>%
    mutate(V6 = gsub('^ +', '', V6)) %>%
    filter(V4 %in% c('D', 'U'))
         
df_kr %>% dfhead
````

````r
# summarizing
p = df_kr %>%
    ggplot(aes(V6, V1)) +
    geom_boxplot() +
    labs(y='% reads mapped') +
    theme_bw()

dims(7,2.5)
plot(p)
````

#### Bracken taxonomy

````r
F = file.path(out_dir, 'kraken', 'all-combined-bracken.tsv')
tax = fread(F, sep='\t')  %>%
    dplyr::select(name, taxonomy_id, taxonomy_lvl, taxonomy, ends_with('_frac')) %>%
    gather(Sample, Perc, -name, -taxonomy_id, -taxonomy_lvl, -taxonomy) %>%
    mutate(Perc = Perc * 100,
           Sample = gsub('_frac$', '', Sample)) %>%
    separate(taxonomy, c('Domain', 'Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species'), sep=';')

tax %>% dfhead
````

````r
tax_s = tax %>%
    group_by(Phylum, Sample) %>%
    summarize(Perc = sum(Perc)) %>%
    ungroup() %>%
    filter(Perc > 0.1, Phylum != '') 
    
tax_s %>% dfhead
````

````r
p = tax_s %>%
    ggplot(aes(Phylum, Perc)) +
    geom_boxplot() +
    scale_y_log10() +
    labs(y='% abundance') +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=90, hjust=1, vjust=0.5)
    )

dims(8, 5)
plot(p)
````

````r
F = file.path(simka_dir, 'mat_presenceAbsence_jaccard.csv.gz')
X = read_mtx(F)
X %>% dfhead
````
