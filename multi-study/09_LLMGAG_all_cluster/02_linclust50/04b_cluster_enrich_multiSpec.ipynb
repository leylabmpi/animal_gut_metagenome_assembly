## KEGG pathway

```r
# KEGG pathways
## Reading file
F = file.path(clust_dir, 'annotate', 'eggnog-mapper', 'clusters_rep-seqs.emapper.annotations.gz')
cmd = glue::glue('gunzip -c {file}', file=F, header=FALSE) 
#cmd = glue::glue('gunzip -c {file} | head -n 10000', file=F, header=FALSE) 
KEGG_ptw = fread(cmd=cmd, sep='\t') 
colnames(KEGG_ptw) = cols
## Formatting
KEGG_ptw = KEGG_ptw %>%
    dt_select(query_name, KEGG_Pathway) %>%
    dt_mutate(KEGG_Pathway = gsub(',map.+', '', KEGG_Pathway))
## Max pathways per seqid 
max_feats = KEGG_ptw %>%
    dt_mutate(n_feats = stringr::str_count(KEGG_Pathway, ',')) %>%
    dt_pull(n_feats) %>% max
new_cols = gsub('^', 'X', 1:(max_feats+1))
## separating 
KEGG_Pathway = KEGG_ptw %>%
    dt_separate(KEGG_Pathway, into=new_cols, sep=',') %>%
    dt_pivot_longer(cols=c(-query_name), names_to='X', values_to='KEGG_pathway') %>%
    dt_select(-X) %>%
    dt_filter(!is.na(KEGG_pathway),
              !grepl('^map', KEGG_pathway)) 
KEGG_Pathway
```

```r
# using kegg ptwy mapping file from humann2 to get pathway hierarchies
ptwy_map_file = '/ebio/abt3_projects/databases_no-backup/humann2/utility_mapping/map_kegg-pwy_name_cat.txt.gz'
cmd = glue::glue('gunzip -c {file}', file=ptwy_map_file)
ptwy_map = fread(cmd=cmd, sep='\t', header=FALSE)
colnames(ptwy_map) = c('KEGG_pathway', 'Level1', 'Level2', 'KEGG_Pathway_Name')
ptwy_map 
```

```r
# joining
KEGG_Pathway = KEGG_Pathway %>%
    dt_left_join(ptwy_map, by=c('KEGG_pathway')) %>%
    dt_rename(seqid = query_name)
KEGG_Pathway
```

```r
# listing files
D = file.path(work_dir, 'multi-spec', 'diamond_blastx')
rpk_files = list_files(D, 'dmnd_RPK.tsv')
rpk_files %>% length
rpk_files %>% head(n=3)
```

```r
# reading in files
read_tbl = function(F){
    sample = basename(dirname(F))
    df = fread(F, sep='\t') %>%
        dt_mutate(frac_abund = ifelse((frac_pos < 0|is.na(frac_pos)), 0, frac_pos)) %>%
        dt_filter(frac_abund > 0) %>%
        dt_select(-slen, -num_reads_hit, -mean_coverage)
    df$Sample = sample
    return(df)
}

doParallel::registerDoParallel(4)
rpk = rpk_files %>%
    #head(n=30) %>%     # DEBUG!!!!!!!!!
    as.list %>%
    plyr::llply(read_tbl, .parallel=TRUE) %>%
    rbindlist(use.names = TRUE) %>%
    dt_mutate(total_abund = sum(frac_abund),
              by=list(sseqid)) %>%
    dt_filter(total_abund > 0) %>%
    dt_select(-total_abund) %>%
    dt_mutate(frac_abund = frac_abund * 100) %>%
    dt_rename(seqid = sseqid,
              perc_abund = frac_abund) 
    
rpk %>% dt_pull(seqid) %>% unique %>% length %>% cat('Number of clusters:', ., '\n')
rpk
```

#### Average number per biome

```r
# counting clusters
rpk_s = rpk %>%
    dt_distinct(Domain, Phylum, seqid, Sample, class) %>%
    dt_mutate(is_mammal = ifelse(class == 'Mammalia', 'MammalTrue', 'MammalFalse')) %>%
    dt_count(by=list(Domain, Phylum, Sample, is_mammal))
rpk_s
```

#### L2FC

```r
# Clusters per phylum + host-class
rpk_s = rpk %>%
    dt_distinct(Domain, Phylum, seqid, Sample) %>%
    dt_count(by=list(Domain, Phylum, Sample))  %>%
    dt_mutate(uuid = dt_map_chr(Phylum, function(x) rep(uuid::UUIDgenerate(), length(x))),
              by=list(Domain, Phylum))

rpk_s %>% dt_pull(N) %>% summary_x('No. of clusters')
rpk_s
```

```r
# checking uuids
rpk_s$Phylum %>% unique_n()
rpk_s$uuid %>% unique_n()
```

```r
# formatting for DESeq2
## feature x observation matrix
countData = rpk_s %>%
    #dt_slice_head(n=1000) %>%  # debug
    dt_pivot_wider(id_cols=uuid, names_from=Sample, values_from=N) %>%
    dt_mutate_if(is.numeric, function(x) ifelse(is.na(x), 0, x)) %>%
    as.data.frame
                 
rownames(countData) = countData$uuid
countData$uuid = NULL

countData
```

```r
# checking key overlap
overlap(countData %>% colnames, colData %>% rownames)
```

```r
# ordering samples by ColData
countData = countData[,colData %>% rownames]
countData
```

```r
# checking prevalence (%)
prev = countData %>%
    apply(1, function(x) sum(x>0) / length(x) * 100) 
prev %>% summary_x('Prevalence')
```

```r
# filtering samples with low prevalence 
prev_cutoff = 50
countData = countData[prev > prev_cutoff,]

# summary
countData
```

```r
# filtering taxa with low total counts
countData %>% rowSums %>% summary_x('Total counts')
countData = countData[rowSums(countData) >= 500,]
countData
```

```r
# design
colData$is_mammal %>% levels
design = ~ is_mammal + Total.Sequences

# creaing obj
dds = DESeqDataSetFromMatrix(countData = countData,
                             colData = colData,
                             design = design)
dds
```

```r
# running DESeq in parallel
register(MulticoreParam(threads))
dds = estimateSizeFactors(dds,  type='poscounts')
dds = DESeq(dds, parallel=TRUE)
```
