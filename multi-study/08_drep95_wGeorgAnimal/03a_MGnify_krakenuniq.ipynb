````r
F = file.path(work_dir, 'krakenuniq', 'report.tsv.gz')
cmd = glue::glue('gunzip -c {file}', file=F)
ku = fread(cmd=cmd, sep='\t') %>%
    as_tibble %>%
    filter(taxID > 1e9)
    
ku
````

````r
# adding sample metadata
ku = ku %>%
    inner_join(MGnify_samp_f %>% 
                   dplyr::select(Sample, C, diversity, Total.Sequences,
                                 biome_1, biome_2, biome_3, biome_4, biome_5),
               by=c('sample'='Sample')) 

ku 
````

# Filtering

## by kmers

````r
ku$sample %>% unique %>% length
ku$taxName %>% unique %>% length
````

````r
# filtering by kmers; if < cutoff, then `perc` = 0
tmp = ku %>%
    mutate(sample = sample %>% factor(levels = unique(sample)),
           taxName = taxName %>% factor(levels = unique(taxName))) %>%
    filter(kmers > 1000) %>%
    dplyr::select(sample, taxName, perc, reads) %>%
    complete(sample, taxName) %>%
    mutate(perc = perc %>% as.Num,
           perc = ifelse(is.na(perc), 0, perc),
           perc = ifelse(perc < 1e-6, 0, perc))

# adding back metadata
## MAG metadata
ku_f = tmp %>%
    inner_join(ku %>% distinct(taxName, Completeness, Contamination,
                               Domain, Phylum, Class, Order, Family, Genus), 
              by=c('taxName')) 
## sample metadata
ku_f = ku_f %>%
    inner_join(ku %>% distinct(sample, biome_1, biome_2, biome_3, biome_4,
                               C, diversity, Total.Sequences), 
               by=c('sample')) 

# status
tmp = NULL
ku_f 
````

````r
# summary
ku_f$sample %>% unique %>% length
ku_f$taxName %>% unique %>% length
````

````r
# filtering out Engineered 
ku_f = ku_f %>%
    filter(biome_1 != 'Engineered') 

ku_f
````

## by prevalence

````r
# calc prev
n_samp = ku_f$sample %>% unique %>% length

prev = ku_f %>% 
    distinct(sample, taxName, perc) %>%
    group_by(taxName) %>%
    summarize(prev = sum(perc > 0) / n_samp * 100) %>%
    ungroup()

prev$prev %>% summary %>% print
prev
````

# Wilcox by biome

* Which MAGs are enriched in host-associated?

````r
# data for all MAGs in all samples?
ku_f %>%
    group_by(taxName) %>%
    summarize(n = n()) %>%
    ungroup() %>%
    .$n %>% summary
````

````r
#' wilcox & output formatting
wilcox_test = function(taxon, df, alt='less'){
    df = df %>% filter(taxName == taxon,
                       biome_1 != 'Engineered')
    res = wilcox.test(x=df[df$biome_1 == 'Environmental',]$perc, 
                      y=df[df$biome_1 == 'Host-associated',]$perc,
                      alternative=alt)    
    df = data.frame(taxName = taxon,
                    statistic = res$statistic,
                    p_value = res$p.value)
    return(df)
}

# selecting taxa
taxa = ku_f$taxName %>% unique %>% as.list
cat('Number of taxa:', taxa %>% length %>% print, '\n')

# running in parallel
doParallel::registerDoParallel(threads)
wilcox_res = plyr::llply(taxa, wilcox_test, df=ku_f, .parallel=TRUE) %>%
    do.call(rbind, .)

# results
wilcox_res
````

````r
# formatting
wilcox_res = wilcox_res %>%
    mutate(p_value = p_value %>% as.Num,
           q_value = p.adjust(p_value, method='BH')) 

wilcox_res$p_value %>% summary
wilcox_res$q_value %>% summary
````

````r
# adding MAG metadata
wilcox_res_j = wilcox_res %>%
    left_join(ku %>% distinct(taxName, Completeness, Contamination, 
                               Domain, Phylum, Class, Order, Family, Genus), 
              c('taxName')) 

wilcox_res_j
````

## Summary

````r
# percent sig
n_taxa = wilcox_res_j %>% nrow
n_sig_taxa = wilcox_res_j %>%
    filter(q_value < 0.01) %>%
    nrow

glue::glue('{x} of {y} MAGs were significant ({z}%)', 
           x=n_sig_taxa, y=n_taxa, 
           z=round(n_sig_taxa / n_taxa * 100, 1))
````

````r
# total summary
p = wilcox_res_j %>%
    mutate(sig = q_value < 0.01, 
           Phylum = gsub('^p__', '', Phylum)) %>%
    group_by(sig, Domain, Phylum) %>%
    summarize(n = n()) %>%
    ungroup() %>%
    ggplot(aes(Phylum, n, fill=sig, label=n)) +
    geom_bar(stat='identity', position='dodge') +
    scale_fill_discrete('Adj. P < 0.01') +
    geom_text(position=position_dodge(width=0.9), 
              vjust=0.5, hjust=-0.2, size=3) +
    scale_y_continuous(limits=c(0, 500), breaks=c(1, 100, 200, 300, 400, 500)) +
    labs(x='Phylum', y='No. of MAGs') +
    coord_flip() +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=55, hjust=1)
    )

dims(5,6)
Plot(p)
````

## Prevalence: all samples

````r
# calculating prev
n_samples = ku_f_j$sample %>% unique %>% length
ku_f_j_s = ku_f_j %>%
    filter(perc > 0) %>%
    group_by(X.BGC.Name, Clan.Number, Family.Number, 
             BGC_type, MAG_BGC, BGC_fam_has_ref,
             Phylum, Class, Order, Family, Genus, taxName) %>%
    summarize(prev = n_distinct(sample) / n_samples * 100) %>%
    ungroup()

ku_f_j_s$prev %>% summary %>% print
ku_f_j_s$prev %>% sd %>% print
ku_f_j_s
````
