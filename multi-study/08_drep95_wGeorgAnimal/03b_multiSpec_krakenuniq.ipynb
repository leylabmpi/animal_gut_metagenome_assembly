## Wilcox by mam vs non-mam

### Mammal vs non-mammal

* similar samples sizes, so less issues with imbalances

````r
# number of samples
dt.dims(20)
ku_f %>%
    group_by(class) %>%
    summarize(n_samples = n_distinct(sample)) %>%
    ungroup()
dt.dims(4)
````

````r
# filtering by prevalence
prev_cutoff = 20
ku_f_f = ku_f %>%
    group_by(class) %>%
    mutate(n_samples = n_distinct(sample)) %>%
    ungroup() %>%
    mutate(present = perc > 1e-5) %>%
    group_by(taxName, Phylum, class) %>%
    summarize(prevalence = sum(present == TRUE) / dplyr::first(n_samples) * 100) %>%
    ungroup() %>%
    filter(prevalence >= prev_cutoff)

cat('Number of MAGs with prev. of >=', prev_cutoff, 
    ':', ku_f_f$taxName %>% unique %>% length, '\n')
ku_f_f
````

````r
# all MAGs (with prev_cutoff)
tax_names = ku_f_f$taxName %>% unique

# kruskal test function
kw_test = function(tax_name, df){
    df = df %>%
        filter(taxName == tax_name) %>%
        mutate(is_mammal = ifelse(class == 'Mammalia', 'mammal', 'Non-mammal')) 
    x = wilcox.test(df %>% filter(class == 'Mammalia') %>% .$perc,
                    df %>% filter(class != 'Mammalia') %>% .$perc,
                    alternative = c("two.sided"))
    c(tax_name, x$statistic, x$p.value)
}

# applying test to all
ku_f_s = tax_names %>%
    as.list %>%
    lapply(kw_test, df=ku_f) %>%
    do.call(rbind, .)

# formatting
colnames(ku_f_s) = c('taxName', 'W', 'p')
ku_f_s = ku_f_s %>%
    as.data.frame %>%
    mutate(p = p %>% as.Num,
           q = p.adjust(p))

# status
ku_f_s$p %>% summary %>% print
ku_f_s$q %>% summary %>% print
ku_f_s 
````

````r
# sig taxa
## 0.05
ku_f_s %>% filter(q < 0.05) %>% nrow %>% print
## 0.01
ku_f_s %>% filter(q < 0.01) %>% nrow %>% print
````

````r
p = ku_f_f_j %>%
    ggplot(aes(Taxonomy, mean_perc, fill=is_mammal)) +
    geom_linerange(aes(ymin=mean_perc,
                       ymax=mean_perc+sd_perc)) +
    geom_bar(stat='identity') +
    scale_fill_discrete('Enrichment') +
    coord_flip() +
    labs(x='Phylum;Class', y='Mean % abund.') +
    theme_bw() +
    theme(
        axis.text.y = element_text(size=8)
    )

dims(6,9,res=220)
plot(p)
````

## DESeq2

### Count matrix

````r
# observation x metadata 
colData = ku_f %>%
    distinct(sample, diet, habitat, class, order, family, genus,
             nonpareil_coverage, nonpareil_diversity, Total.Sequences, 
             scientific_name) %>%
    mutate(is_mammal = ifelse(class == 'Mammalia', 'MammalTrue', 'MammalFalse'),
           is_mammal = is_mammal %>% as.factor,
            Total.Sequences = Total.Sequences %>% scale) %>%
    as.data.frame

rownames(colData) = colData$sample
colData$sample = NULL

# order df
colData = colData[colnames(countData),]

# status
colData
````

## Prevalence

````r
# filtering by kmers; if < cutoff, then `perc` = 0
## not filtering by prevalence (as used for DESeq2)
tmp = ku %>%
    mutate(sample = sample %>% factor(levels = unique(sample)),
           taxName = taxName %>% factor(levels = unique(taxName))) %>%
    filter(kmers > 1000) %>%
    dplyr::select(sample, taxName, perc, reads) %>%
    complete(sample, taxName) %>%
    mutate(perc = perc %>% as.Num,
           perc = ifelse(is.na(perc), 0, perc),
           perc = ifelse(perc < 1e-6, 0, perc))

# adding back metadata
## MAG metadata
ku_f = tmp %>%
    inner_join(ku %>% distinct(taxName, Completeness, Contamination,
                               Domain, Phylum, Class, Order, Family, Genus), 
              c('taxName')) 
## sample metadata
ku_f = ku_f %>%
    inner_join(ku %>% distinct(sample, nonpareil_coverage, nonpareil_diversity, 
                               Total.Sequences, class, order, family, genus,
                               scientific_name, diet, habitat), 
               c('sample')) 

# status
tmp = NULL
cat('No. of MAGs:', ku_f$taxName %>% unique %>% length, '\n')
ku_f 
````

````r
# prevalence of each MAG across all animals
ku_f_s = ku_f %>%
    mutate(n_samples = n_distinct(sample)) %>%
    mutate(present = perc > 1e-5) %>%
    group_by(taxName) %>%
    summarize(prevalence = sum(present == TRUE) / dplyr::first(n_samples) * 100) %>%
    ungroup()

ku_f_s$taxName %>% unique %>% length
ku_f_s$prevalence %>% summary %>% print
ku_f_s
````

````r
# writing table
F = file.path(work_dir, 'MAG_KU-multiHost-prev.tsv')
write_table(ku_f_s, F)
````

````r
# prevalence of each MAG across all animals
## per class
ku_f_s = ku_f %>%
    group_by(class) %>%
    mutate(n_samples = n_distinct(sample)) %>%
    ungroup() %>%
    mutate(present = perc > 1e-5) %>%
    group_by(taxName, class) %>%
    summarize(prevalence = sum(present == TRUE) / dplyr::first(n_samples) * 100) %>%
    ungroup() %>%
    spread(class, prevalence)

ku_f_s
````

````r
# writing table
F = file.path(work_dir, 'MAG_KU-multiHost-prev_byClass.tsv')
write_table(ku_f_s, F)
````
