## Taxonomic novelty

* hits to reps in the GTDBr89

````r
GTDBTk_res_file = '/ebio/abt3_projects/Georg_animal_feces/data//metagenome/multi-study/BioProjects/summary/LLMGA/wGeorgAnimal/drep-0.95/drep-MAG_metadata.tsv'

GTDBTk_res = read.delim(GTDBTk_res_file, sep='\t') %>%
    mutate(binID = gsub('_+', '.', binID),
           binID = gsub('\\.+', '.', binID)) 
GTDBTk_res %>% nrow
GTDBTk_res %>% colnames
````

````r
# 95% ANI match
GTDBTk_res_f = GTDBTk_res %>%
    mutate(closest_placement_ani = closest_placement_ani %>% as.character,
           closest_placement_ani = ifelse(closest_placement_ani == 'N/A', 
                                          0, closest_placement_ani),
           closest_placement_ani = ifelse(is.na(closest_placement_ani), 
                                          0, closest_placement_ani),
           closest_placement_ani = closest_placement_ani %>% as.Num) %>%
    mutate(has_species_placement = closest_placement_ani >= 95,
           has_genus_placement = Genus != 'g__', 
           has_family_placement = Family != 'f__')

cat('Number of MAGs with closests match ANI of <95%:',
GTDBTk_res_f %>% filter(closest_placement_ani < 95) %>% nrow, '\n')

cat('Number of MAGs with no genus classification:',
GTDBTk_res_f %>% filter(has_genus_placement == FALSE) %>% nrow, '\n')

cat('Number of MAGs with no family classification:',
GTDBTk_res_f %>% filter(has_family_placement == FALSE) %>% nrow, '\n')
````

````r
# colorstrip
x = GTDBTk_res_f %>%
    dplyr::select(binID, has_species_placement, 
                  has_genus_placement, has_family_placement) %>%
    mutate(novelty = case_when(
           has_family_placement == FALSE ~ 'no_family_match',
           has_genus_placement == FALSE ~ 'no_genus_match',
           has_species_placement == FALSE ~ 'no_species_match',
           has_species_placement == TRUE ~ 'species_match')) %>%
    dplyr::select(binID, novelty) %>%
    as.data.frame

rownames(x) = x$binID
x$binID = NULL

out_file = file.path(phylo_dir, 'itol-clrStrip_GTDB-novelty.txt')
itol_colorstrip(x, 'GTDB-novelty', out_file)
````

## MGnify krakenuniq-DESeq2 data

````r
# Reading in DESeq2 data
MGnify_ku_file = '/ebio/abt3_projects/Georg_animal_feces/data/metagenome/multi-study/BioProjects/summary/LLMGA/wGeorgAnimal/drep-0.95/MGnify-host-env/MGnify-host-env_krkunq_DEseq2.tsv'
MGnify_ku = read.delim(MGnify_ku_file, sep='\t') %>%
    mutate(taxName = gsub('_+', '.', taxName),
           taxName = gsub('\\.+', '.', taxName)) %>%
    dplyr::select(taxName, baseMean, log2FoldChange, lfcSE, stat, pvalue, padj)
MGnify_ku
````

````r
bigscape_dir = file.path(dirname(work_dir), 'BGCs', 'bigscape')
clan_file = file.path(bigscape_dir, 'bigscape_fam-clan.tsv')

clans = read.delim(clan_file, sep='\t') %>%
     mutate(MAG = gsub('_+', '.', MAG),
            MAG = gsub('$', '.contigs', MAG),
            MAG = gsub('\\.+', '.', MAG))
clans
````

````r
# joining 
MAG_meta_f_j = MAG_meta_f %>%
    inner_join(clans, by=c('binID'='MAG'))

MAG_meta_f_j
````

````r
# 95% ANI match
MAG_meta_f_j = MAG_meta_f_j %>%
    mutate(closest_placement_ani = closest_placement_ani %>% as.character,
           closest_placement_ani = ifelse(closest_placement_ani == 'N/A', 
                                          0, closest_placement_ani),
           closest_placement_ani = ifelse(is.na(closest_placement_ani), 
                                          0, closest_placement_ani),
           closest_placement_ani = closest_placement_ani %>% as.Num) %>%
    mutate(has_species_placement = closest_placement_ani >= 95,
           has_genus_placement = Genus != 'g__', 
           has_family_placement = Family != 'f__') %>%
    mutate(novelty = case_when(
           has_family_placement == FALSE ~ 'no_family_match',
           has_genus_placement == FALSE ~ 'no_genus_match',
           has_species_placement == FALSE ~ 'no_species_match',
           has_species_placement == TRUE ~ 'species_match'))

MAG_meta_f_j
````

````r
BGC_counts = MAG_meta_f_j %>%
    group_by(binID) %>%
    summarize(n_BGCs = n_distinct(X.BGC.Name)) %>%
    ungroup()

BGC_counts  = MAG_meta_f_j %>%
    distinct(binID, novelty) %>%
    inner_join(BGC_counts, c('binID')) 

p = BGC_counts %>%
    ggplot(aes(novelty, n_BGCs)) +
    geom_boxplot() +
    labs(x='GTDB novelty', y='No. of BGCs') +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )

p.dims(4, 3)
plot(p)
````

````r
cat('All SBGs\n')
x = BGC_counts$binID %>% unique %>% length
BGC_counts %>%
    group_by(novelty) %>%
    summarize(n_MAGs = n_distinct(binID),
              perc_MAGs = n_distinct(binID) / x * 100,
              mean_BGCs = mean(n_BGCs),
              sd_BGCs = sd(n_BGCs))

cat('BGC >=3\n')
x = BGC_counts %>% filter(n_BGCs >= 3) %>% .$binID %>% unique %>% length
BGC_counts %>%
    filter(n_BGCs >= 3) %>%
    group_by(novelty) %>%
    summarize(n_MAGs = n_distinct(binID),
              perc_MAGs = n_distinct(binID) / x * 100,
              mean_BGCs = mean(n_BGCs),
              sd_BGCs = sd(n_BGCs))
````

### Bootstrap p-value

````r
# permuting "sig" labels among MAGs
perm_labels = function(rep, df){
    psbl_sigs = c('Host', 'Environment', 'Neither')
    df = df %>%
        group_by(taxName) %>%
        mutate(sig = sample(psbl_sigs, 1)) %>%
        group_by(sig, phenotype, phenotype_model) %>%
        summarize(perc_MAGs_with_pheno = sum(prediction) / n() * 100) %>%
        ungroup() %>%
        mutate(sim_rep = rep) 
    return(df)
}
````

````r
# formatting
tmp = trt_maj_j %>%
    filter(!is.na(phenotype)) %>%
    dplyr::select(taxName, phenotype, prediction, phenotype_model, sig, padj) %>%
    mutate(prediction = prediction %>% as.Num)
# ground truth: percent of MAGs with phenotype in "sig"
truth = tmp %>%
    group_by(sig, phenotype, phenotype_model) %>%
    summarize(perc_MAGs_with_pheno = sum(prediction) / n() * 100) %>%
    ungroup() %>%
    mutate(sim_rep = 0) 

# permuting the data & comparing to the ground truth
perm_reps = 1:1000
trt_maj_j_s = perm_reps %>%
    as.list %>% lapply(perm_labels, df=tmp) %>%
    do.call(rbind, .) %>%
    inner_join(truth, c('sig', 'phenotype', 'phenotype_model')) %>%
    group_by(sig, phenotype, phenotype_model) %>%
    summarize(perc_gt_in_truth = sum(perc_MAGs_with_pheno.y > perc_MAGs_with_pheno.x) / 
                                     length(perm_reps) * 100) %>%
    ungroup() %>%
    mutate(boot_p = (100 - perc_gt_in_truth) / 100)

# summary
tmp = truth = NULL
trt_maj_j_s$perc_gt_in_truth %>% summary
trt_maj_j_s$boot_p %>% summary
trt_maj_j_s
````

````r
# more detailed summary
trt_maj_j_s_f = trt_maj_j_s %>%
    filter(phenotype_model == 'phypat+PGL', 
           boot_p < 0.05) 
    
trt_maj_j_s_f$phenotype %>% unique_n('phenotypes')
trt_maj_j_s_f %>%
    group_by(sig) %>%
    summarize(n_pheno = n_distinct(phenotype)) %>%
    ungroup()
````

````r
# formatting
trt_maj_j_j = trt_maj_j %>%
    filter(!is.na(phenotype),
           phenotype_model == 'phypat+PGL') %>%
    mutate(prediction = as.Num(prediction) == 1) %>%
    group_by(phenotype, prediction, sig) %>%
    summarize(n = n()) %>%
    ungroup() %>%
    group_by(phenotype) %>%
    mutate(pos_pred = max(ifelse(prediction == TRUE, n, NA), na.rm=TRUE)) %>%
    ungroup() %>%
    left_join(trt_maj_s_f, c('sig', 'phenotype')) 

trt_maj_j_j
````

````r
# plotting
p = trt_maj_j_j %>%
    mutate(phenotype = phenotype %>% reorder(pos_pred)) %>%
    ggplot(aes(phenotype, n, fill=prediction)) +
    geom_bar(stat='identity') +
    geom_text(data=trt_maj_j_j %>% filter(!is.na(boot_p), boot_p < 0.01, prediction==TRUE),
              aes(label='*', y=n), vjust=0.8) +
    scale_fill_discrete('Phenotype\npredicted?') +
    facet_wrap(~sig, scales='free_x') +
    labs(y='No. of MAGs', x='Phenotype') +
    coord_flip() +
    theme_bw() +
    theme(
        axis.text.y = element_text(size=8)
    )

p.dims(7,7)
Plot(p)
````
